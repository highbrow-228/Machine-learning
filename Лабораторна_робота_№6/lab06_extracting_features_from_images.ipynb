## [Dataset](https://www.kaggle.com/datasets/andrewmvd/face-mask-detection/data)

```python
import matplotlib.pyplot as plt
from skimage.io import imread

# Завантаження зображення (вибираємо з набору даних)
image = imread("..\\data\\dataset_for_lab06\\images\\maksssksksss25.png")

# Виведення інформації про розмір зображення та візуалізація
print(f"Розмір зображення: {image.shape}")
plt.imshow(image)
plt.show()
```

```python
from skimage.color import rgb2gray

# Якщо зображення кольорове, переводимо в відтінки сірого
if len(image.shape) == 3:
    # Відкидаємо альфа-канал, якщо він є
    image_rgb = image[:, :, :3]
    gray_image = rgb2gray(image_rgb)
else:
    gray_image = image

# Перевірка результату
print(f"Shape of grayscale image: {gray_image.shape}")
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
# Вилучення піксельних значень
pixel_features = gray_image.flatten()

# Перевірка кількості ознак
print(f"Кількість характеристик пікселів: {pixel_features.shape[0]}")
```

```python
import numpy as np

if len(image.shape) == 3:
    mean_red = np.mean(image[:, :, 0])
    mean_green = np.mean(image[:, :, 1])
    mean_blue = np.mean(image[:, :, 2])
    print(f"Середні значення для каналів RGB:\n\tRed = {mean_red},\n\tGreen = {mean_green},\n\tBlue = {mean_blue}")
else:
    print("Зображення у відтінках сірого, тому немає необхідності обчислювати середнє значення RGB.")
```

```python
from skimage.filters import prewitt

# Використання ядра Prewitt для виявлення контурів
edges = prewitt(gray_image)

# Виведення результату
plt.imshow(edges, cmap='gray')
plt.show()
```

```python
print("Довжина елементів на основі пікселів:", len(pixel_features))
print("Довжина елементів на основі ребер:", edges.flatten().shape[0])
print(f"Середня довжина RGB-функцій: червоний ({mean_red}), зелений({mean_green}), синій({mean_blue})")
```
