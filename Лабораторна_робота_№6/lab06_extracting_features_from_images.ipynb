## [Dataset](https://www.kaggle.com/datasets/andrewmvd/face-mask-detection/data)

```python
import matplotlib.pyplot as plt
from skimage.io import imread

#   (   )
image = imread("..\\data\\dataset_for_lab06\\images\\maksssksksss25.png")

#       
print(f" : {image.shape}")
plt.imshow(image)
plt.show()
```

```python
from skimage.color import rgb2gray

#   ,    
if len(image.shape) == 3:
    #  -,   
    image_rgb = image[:, :, :3]
    gray_image = rgb2gray(image_rgb)
else:
    gray_image = image

#  
print(f"Shape of grayscale image: {gray_image.shape}")
plt.imshow(gray_image, cmap='gray')
plt.show()
```

```python
#   
pixel_features = gray_image.flatten()

#   
print(f"  : {pixel_features.shape[0]}")
```

```python
import numpy as np

if len(image.shape) == 3:
    mean_red = np.mean(image[:, :, 0])
    mean_green = np.mean(image[:, :, 1])
    mean_blue = np.mean(image[:, :, 2])
    print(f"    RGB:\n\tRed = {mean_red},\n\tGreen = {mean_green},\n\tBlue = {mean_blue}")
else:
    print("   ,       RGB.")
```

```python
from skimage.filters import prewitt

#   Prewitt   
edges = prewitt(gray_image)

#  
plt.imshow(edges, cmap='gray')
plt.show()
```

```python
print("    :", len(pixel_features))
print("    :", edges.flatten().shape[0])
print(f"  RGB-:  ({mean_red}), ({mean_green}), ({mean_blue})")
```
